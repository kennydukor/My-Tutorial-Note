# Simple Linear Regression with Auto Insurance Dataset Using Least Square Method

## This first method will build a linear regression using libraries and not from scratch

IMPORT LIBRARIES

```python
import pandas as pd    # this is used to manage the csv and excel files.
import matplotlib.pyplot as plt    # this is responsible for plotting of the data
from statsmodels import api as sm
import seaborn as sns  # This will make the graphs pretty
sns.set()
```

LOAD YOUR DATA

```python
url = 'https://github.com/kennydukor/Datasets/raw/master/auto-insurance.csv' # this will store dataset link in the variable 'url'
data = pd.read_csv(url, names = ["NoOfClaims", "Payment"]) # pandas will read the datasdt because its in csv format
```

```python
data  # display the csv file
```

```python
data.describe()   # this will give a statisical description of the dataset
```

### VISUALIZE DATA

```python
y = data['Payment']
x1 = data['NoOfClaims']
```

```python
plt.scatter(x1, y)
plt.xlabel('Number of Claims')
plt.ylabel('Payment')
plt.show()
```

### Generate Regression Data

```python
x = sm.add_constant(x1) # this is trying to ensure that theta0 does not stand alone. therefore, we add x = 1
#ie y = (x * theta0) + (x1 * theta1)
result = sm.OLS(y,x).fit()   #the ordinary least square regression (OLS)
result.summary()
```

### PLOT REGRESSION LINE

```python
plt.scatter(x1, y)
yhat = 3.4138*x1 + 19.9945
fig = plt.plot(x1, yhat, lw =4, c ='orange', label ='regression line')
plt.xlabel('Number of Claims')
plt.ylabel('Payment')
plt.show()
```

### MAKE PREDICTION

```python
x1 = float(input("No of Claims to predict the amount to be paid: "))
yhat = 3.4138*x1 + 19.9945
print(yhat)
```
